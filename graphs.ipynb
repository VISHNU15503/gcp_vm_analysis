```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv('VM_measuremnets.csv')
```

```python
# Show all different instance types
print(df['instance_type'].unique())
```

```python
# remove trail_no column
df = df.drop('trail_no', axis=1)
```

```python
# Show the average cold_process_time, warm_process_time, running_process_time for each instance type and remove trail numbers
df = df.groupby(['instance_type']).mean()
```

```python
df
```

```python
# order df by e2-small,e2-medium,n1-standard-2,n1-standard-4,n2-standard-2,n2-standard-4,n2d-standard-2,n2d-standard-4,e2-standard-2,e2-standard-4
df = df.reindex(['e2-small','e2-medium','n1-standard-2','n1-standard-4','n2-standard-2','n2-standard-4','n2d-standard-2','n2d-standard-4','e2-standard-2','e2-standard-4'])
```

```python
df
```

```python
# Plot the bar graph having x-axis as instance type and y-axis as average cold_process_time
plt.figure(figsize=(15,5))
plt.bar(df.index, df['cold_process_time'], color='royalblue',width=0.5)
plt.plot(df.index, df['cold_process_time'], color='red',marker='o')
plt.xlabel('Instance Type')
plt.ylabel('Average Cold Process Time (in seconds)')
plt.title('Average Cold Process Time for different Instance Types')
plt.show()
```

```python
# Plot the bar graph having x-axis as instance type and y-axis as average warm_process_time	
plt.figure(figsize=(15,5))
plt.bar(df.index, df['warm_process_time'], color='royalblue',width=0.5)
plt.plot(df.index, df['warm_process_time'], color='darkorange',marker='o')
plt.xlabel('Instance Type')
plt.ylabel('Average Warm Process Time (in seconds)')
plt.title('Average Warm Process Time for different Instance Types')
plt.show()
```

```python
# Plot the bar graph having x-axis as instance type and y-axis as average running_process_time	
plt.figure(figsize=(15,5))
plt.bar(df.index, df['running_process_time'], color='royalblue',width=0.5)
plt.plot(df.index, df['running_process_time'], color='magenta',marker='o')
plt.xlabel('Instance Type')
plt.ylabel('Average Running Process Time (in seconds)')
plt.title('Average Running Process Time for different Instance Types')
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot(df.index, df['cold_process_time'], color='red',marker='o')
plt.plot(df.index, df['warm_process_time'], color='darkorange',marker='o')
plt.plot(df.index, df['running_process_time'], color='magenta',marker='o')
plt.legend(['Cold Process Time','Warm Process Time','Running Process Time'])
plt.xlabel('Instance Type')
plt.ylabel('Process Times (in seconds)')
plt.title('Process Time for different Instance Types')
plt.show()
```
